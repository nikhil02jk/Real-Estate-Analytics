```python
import pandas as pd
```

```python
df1=pd.read_csv('API_DATA.csv')
```

```python
df1.info()
```

```python
df1.head()
```

```python
df1=df1.drop(columns=["carouselPhotos","detailUrl","contingentListingType","imgSrc","currency","hasImage","listingSubType"])
```

```python
df1.head()
```

```python
import numpy as np

mask = df1['address'].str.contains('zpid/', na=False)

# Set 'address' to NaN for rows where 'zpid/' is found
df1.loc[mask, 'address'] = np.nan

# Shift columns starting from 'bedrooms' to the right by one column
df1.loc[mask, 'bedrooms':] = df1.loc[mask, 'bedrooms':].shift(1, axis=1)
```

```python
df1["has3DModel"].unique()
```

```python
df1.head()
```

```python
# Ensure all values in 'has3DModel' are strings and lowercase
df1['has3DModel'] = df1['has3DModel'].astype(str).str.lower()
```

```python
df1.head()
```

```python
df1["has3DModel"].unique()
```

```python
# Map 'true' and 'false' to 1 and 0 respectively, handle 'nan' as 0
df1['has3DModel'] = df1['has3DModel'].map({'true': 1, 'false': 0}).fillna(0).astype(int)
```

```python
df1["has3DModel"].unique()
```

```python
df1.head()
```

```python
df2=df1.copy()
```

```python
df2.head()
```

```python
df2=df2.drop(columns=["Unnamed: 0","listingStatus","country","variableData"])
```

```python
df2.head()
```

```python
# Define the IQR method for outlier removal
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply outlier removal
df_cleaned = remove_outliers_iqr(df2, 'daysOnZillow')
```

```python
df_cleaned['daysOnZillow'] = df_cleaned['daysOnZillow'].astype(int)
```

```python
df_cleaned.head()
```

```python
df_cleaned[['Street Address', 'City', 'State/ZIP']] = df_cleaned['address'].str.extract(r'^(.*),\s(.*),\s([A-Z]{2}\s\d{5})$')
```

```python
df_cleaned.head()
```

```python
df_cleaned['unit'] = df_cleaned['unit'].str.replace('#', '').str.strip()
```

```python
df_cleaned=df_cleaned.drop(columns=["address"])
```

```python
df_cleaned.to_csv("API_DATA4.csv")
```

```python
data=df_cleaned.copy()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Distribution of Bathrooms and Bedrooms
plt.figure(figsize=(10, 6))
sns.histplot(data['bathrooms'], kde=False, bins=10, color='blue', label='Bathrooms')
sns.histplot(data['bedrooms'], kde=False, bins=10, color='orange', label='Bedrooms', alpha=0.7)
plt.title('Distribution of Bathrooms and Bedrooms')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# 2. Price Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], kde=True, bins=30, color='green')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='livingArea', y='price', data=data, alpha=0.7, hue='City')
plt.title('Living Area vs Price')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Price')
plt.legend(title='City')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['City'].value_counts().plot(kind='bar', color='purple')
plt.title('Number of Properties by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=data)
plt.title('Price vs Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
data['has3DModel'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title('Properties with and without 3D Models')
plt.xlabel('3D Model Availability (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['daysOnZillow'], kde=True, bins=30, color='purple')
plt.title('Days on Zillow Distribution')
plt.xlabel('Days on Zillow')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
numerical_columns = ['bathrooms', 'bedrooms', 'price', 'zestimate', 'daysOnZillow', 'rentZestimate', 'livingArea']
correlation_matrix = data[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
