```python
# Step 2: Identify and handle missing values
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
# 1. Handle Missing Data
# Fill missing numerical values with median
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
# Fill missing categorical values with mode
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

```python
# Drop columns with excessive missing values (e.g., > 50%)
missing_threshold = 0.5
missing_percent = df.isnull().mean()
cols_to_drop = missing_percent[missing_percent > missing_threshold].index
df = df.drop(columns=cols_to_drop)
df = df.drop(columns=cols_to_drop)
```

```python
# Dropping rows with missing values
df = df.dropna()
```

```python
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
# 2. Convert Data Types
# Convert datePostedString to datetime
df['datePostedString'] = pd.to_datetime(df['datePostedString'], errors='coerce')
```

```python
# 3. Handle Duplicates
df = df.drop_duplicates()
```

```python
# Ensure longitude is between -180 and 180, latitude is between -90 and 90
df.loc[(df['longitude'] < -180) | (df['longitude'] > 180), 'longitude'] = np.nan
df.loc[(df['latitude'] < -90) | (df['latitude'] > 90), 'latitude'] = np.nan
# Fill missing values with median
df['longitude'] = df['longitude'].fillna(df['longitude'].median())
df['latitude'] = df['latitude'].fillna(df['latitude'].median())
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['bathrooms'].quantile(0.25)
Q3 = df['bathrooms'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_cleaned = df[(df['bathrooms'] >= lower_bound) & (df['bathrooms'] <= upper_bound)]

# Print summary after outlier removal
print("Summary after removing outliers:")
print(df_cleaned['bathrooms'].describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is the DataFrame containing the data
# Convert 'datePostedString' to datetime
df['datePostedString'] = pd.to_datetime(df['datePostedString'], errors='coerce')

# Drop rows with invalid or missing dates
df = df.dropna(subset=['datePostedString'])

# Create a new column for Year-Month
df['Year-Month'] = df['datePostedString'].dt.to_period('M').astype(str)

# Calculate median price for each Year-Month
median_prices = df.groupby('Year-Month')['price'].median().reset_index()

# Plot the price trend
plt.figure(figsize=(12, 6))
plt.plot(median_prices['Year-Month'], median_prices['price'], color='brown', marker='o', linewidth=2)
plt.title("House Price Trends Over Time", fontsize=16)
plt.xlabel("Year-Month", fontsize=12)
plt.ylabel("Median Price ($)", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(visible=True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Combine all descriptions into a single string
text = " ".join(data['description'].dropna().astype(str))  # Remove NaN and ensure all are strings

# Define stopwords
stopwords = set(STOPWORDS)

# Generate the word cloud
wordcloud = WordCloud(
    width=800, height=400,
    background_color='white',
    stopwords=stopwords,
    colormap='viridis',
    max_words=200
).generate(text)

# Plot the word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Property Descriptions", fontsize=16)
plt.show()
```
